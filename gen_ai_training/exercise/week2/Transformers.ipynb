```python
from transformers import GPT2TokenizerFast, GPT2LMHeadModel
```

```python
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2-large")
```

```python
model = GPT2LMHeadModel.from_pretrained("gpt2-large", pad_token_id=tokenizer.eos_token_id)
```

```python
tokenizer.eos_token_id
```

```python
tokenizer.decode(tokenizer.eos_token_id)
```

```python
sentence = "I would like to go to visit the park in the morning"
```

```python
tokens = tokenizer(sentence, return_tensors="pt")
```

```python
tokens
```

```python
tokenizer.decode(tokens['input_ids'][0][1])
```

```python
result = model.generate(tokens['input_ids'], max_length=100, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
print(tokenizer.decode(result[0], skip_special_tokens=True))
```
